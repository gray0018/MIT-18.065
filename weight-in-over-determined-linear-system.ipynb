```python
import numpy as np
```

```python
A = np.random.randint(1000, size=(100, 100))
b = np.random.randint(1000, size=(100, 1))

lam = 10000000
A_append = np.zeros((3,100), dtype=int)
A_append[0, 0] = lam
A_append[1, 1] = lam
A_append[2, 2] = lam

b_append = np.zeros((3,1), dtype=int)

b_append[0, 0] = lam*10
b_append[1, 0] = lam*20
b_append[2, 0] = lam*30

A_prime = np.vstack((A, A_append))
b_prime = np.vstack((b, b_append))



np.hstack((np.linalg.inv(A) @ b, np.linalg.lstsq(A, b, rcond=None)[0], np.linalg.lstsq(A_prime, b_prime, rcond=None)[0]))
```
